```
#load weekly data from DA parquet files to delta table partitioned by week taken from filename
def load_da_weekly_parquet_to_delta_weekly(input, output):
       
        files=glob.glob(deploy_parallel_s + input, recursive=True)
       
        files_week=[e[-14:] for e in files]
        files_week=[e[:-8] for e in files_week]
        print(input + ": ")
        print("files : ",files,"files_week :",files_week)
        
        files=[fi.replace('/dbfs','dbfs:') for fi in files]
        pdldf_all = spark.read.parquet(input_path + input)
        pdldf_all = pdldf_all.withColumn("week_id", f.lit(files_week[0]))
        pdldf_all.write.partitionBy('week_id').format('delta').mode('overwrite').option('replaceWhere',f'week_id = ({files_week[0]})').save(pdldir + output)
```

```
#load monthly data from DA parquet files to delta table partitioned by month
def load_da_monthly_parquet_to_delta_monthly(mnth_start, mnth_end, input, output):
    pdldf_all = spark.read.parquet(input_path + input).filter(f'month_no >= ({mnth_start}) AND month_no <= ({mnth_end})')
    pdldf_all = pdldf_all.withColumn("month_no",f.concat(f.year(pdldf_all.txn_date),f.lpad(f.month(pdldf_all.txn_date),2,'0')))
    pdldf_all.write.partitionBy('month_no').format('delta').mode('overwrite').option('replaceWhere',f'month_no >= ({mnth_start}) AND month_no <= ({mnth_end})').save(pdldir + output)
```
